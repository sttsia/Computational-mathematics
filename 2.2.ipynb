***Американские горки***

Реализуя ручками методы дихотомии и Фиббоначи (золотого сечения), найти точку локального минимума функций (любых двух из _а-ж_ на ваш выбор):

a) $f(x)=2 x^2-\ln x$

б) $f(t)=t^3 / 3+t^2$

в) $f(t)=t^4 / 4-2 t^2$

г) $f(t)=t e^{-t^2 / 2}$

д) $f(t)=3 t^4-8 t^3+6 t^2$

e) $f(t)=(t-5) e^t$

ж) $f(t)=\left(t^2-3\right) /(t+2)$.

```python
import numpy as np

def f1(x):
  return 2*x**2-np.log(x)

def f2(x):
  return x**3/3+x**2

eps = 1e-6
a = 0
b = 1
while (b-a)<eps:
  c = (b-a)/2
  f_left = f1(c-eps)
  f_right = f1(c+eps)
  if f_left < f_right:
    b = c
  else:
    a = c
print('Минимум f1 (дихотомия): ',(a+b)/2)

a = -1
b = 0.5
N = 100
fib = [1,1]
for i in range(2,N+1):
  fib.append(fib[i-1]+fib[i-2])
x1 = a+(b-a)*fib[N-2]/fib[N]
x2 = a+(b-a)*fib[N-1]/fib[N]
f_1 = f2(x1)
f_2 = f2(x2)
for i in range(N,0,-1):
  if (x2-x1)<1e-8:
    break
  if f_1 > f_2:
    a = x1
    x1 = x2
    x2 = b-(x1-a)
    f_1 = f_2
    f_2 = f2(x2)
  else:
    b = x2
    x2 = x1
    x1 = a+(b-x2)
    f_2 = f_1
    f_1 = f2(x1)
print('Минимум f2 (Фибоначчи): ',(x1+x2)/2)
```

***Функция Растригина***

$$
f\left(x_1, x_2\right)=20+x_1^2+x_2^2-10 \cos \left(2 \pi x_1\right)-10 \cos \left(2 \pi x_2\right)
$$

Постройте множество локальных минимумов и максимумов на плоскости используя `minimize`. Минимумы должны быть синими, максимумы - красными. В 3D график функции выглядит примерно так:

```python
from scipy.optimize import minimize, Bounds
import numpy as np
import matplotlib.pyplot as plt
from math import pi

def func(x):
  return 20 + x[0]**2 + x[1]**2 - 10*np.cos(2*pi*x[0]) - 10*np.cos(2*pi*x[1])

def func1(x):
  return (-1)*(20 + x[0]**2 + x[1]**2 - 10*np.cos(2*pi*x[0]) - 10*np.cos(2*pi*x[1]))

mins_x = []
mins_y = []
maxs_x = []
maxs_y = []
for i in range(-5,6,1):
  for j in range(-5,6,1):
    bounds = Bounds ([i-0.5, j-0.5], [i+0.5, j+0.5])
    mins = minimize(func,(i,j),bounds=bounds)['x']
    mins_x.append(mins[0])
    mins_y.append(mins[1])
for i in range(-5,5,1):
  for j in range(-5,5,1):
    bounds = Bounds ([i, j], [i+1, j+1])
    maxs = minimize(func1,(i+0.5,j+0.5),bounds=bounds)['x']
    maxs_x.append(maxs[0])
    maxs_y.append(maxs[1])

plt.scatter(mins_x,mins_y)
plt.scatter(maxs_x,maxs_y,color='r')
```

***Градиенты градиентами***

Первая картинка относится к методу наискорейшего спуска (тк шаги ортогональны). Соостветственно вторая картинка относится к методу сопряженных градиентов. Замкнутые линии показывают, где значение функции одинаковое.

В случае квадратичной функции $\mathbb {R} ^{n}$ минимум находится не более чем за $n$ шагов.

__Док-во:__

Пусть надо оптимизировать $F(x) = \frac{1}{2}(Ax,x)-(b,x) \rightarrow inf, \ x \in \mathbb{R}^n.$

Тогда нахождение минимума эквивалентно $F'(x)=0,$ т.е. системе $Ax-b=0$. Решение системы раскладывается по базису
$$x^* = x_0+a_1p_1+...+a_np_n,$$
где $x_0$ - произвольная точка из $\mathbb {R} ^{n}$ (начальная в методе сопряженных градиентов).

Согласно методу ищется разложение по базисным векторам, причем на каждом шаге добавляется один базисный вектор. Поэтому количество итераций не будет превышать $n$ (оно будет меньше, если в разложении вектора $x^*$ присутствуют не все базисные вектора)

***Овальный***


Если функция овражная, то, когда алгоритм ищет минимум вдоль градиента, он будет проходить малое расстояние. Чтобы дойти до минимума (как на рисунке), нужно пройти вдоль внутреннего овала (т.к. так работает метод), а это займет много итераций, потому что вектора в данном методе получаются перпендикулярными.

Если аппроксимировать функцию около минимума квадратичной формой, то наименьшее сингулярное число будет значительно меньше других, так как по условию линии сильно вытянуты в одном направлении. Это направление и будет соответствовать наименьшему сингулярному числу.

При взятии обратной матрицы из наименьшего сингулярного числа получится наибольшее, поэтому наибольшее сингулярное число $||A^{-1}||$ будет существенно больше других сингулярных чисел. При этом, так как $||A||=\sqrt{\max \lambda}$, то получится, что число обусловленности $K = ||A|| \cdot ||A^{-1}|| \gg 1$.

Вывод из рисунка про метод координатного спуска: для овражных функций он тоже работает плохо, поэтому с такими функциями работать тяжело.

***Сопрягаемся (полузвёздочка)***

Рассмотрим на примере функции Стайблински-Танга поведение различных вариантов метода сопряженных градиентов. Функция задается выражением
$$
f(x)=\frac{1}{2} \sum_{i=1}^n x_i^4-16 x_i^2+5 x_i .
$$
Глобальным минимум достигается в точке $(\eta, \ldots, \eta)$, где $\eta \approx-2.904$. Использовать данную функцию будем при $n=2$, чтобы наглядно проиллюстрировать результаты.

Напишите все 7 вариантов __(*, если не на доп. быллы, то возьмите 2 из алгоритма)__ реализации метода сопряженных градиентов из семинара. Сравните результаты. Функция специально подобрана таким образом, что некоторые из вариантов реализации ответ не смогут получить, а остальные продемонстрируют разное качество работы.

Качество измерьте как количество итераций, необходимое для получения точности

$$
\left\|r\right\|<10^{-6}
$$

Пусть метод останавливается на итерации $k$, если выполняется условие $\left\|r_k\right\|<10^{-6}$, или по истечению 1000 итераций.

Начальные приближения берите следующие:

(0.1, -1)

(0.1, -1.1)

(0, -1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def f(x):
    return (1/2)*(x[0]**4-16*x[0]**2+5*x[0]+x[1]**4-16*x[1]**2+5*x[1])

def grad_f(x):
    return np.array([(1/2)*(4*x[0]**3-32*x[0]+5), (1/2)*(4*x[1]**3-32*x[1]+5)])

def conjugate_gradients(starting_point, epsilon, param):
    x = starting_point
    iterations = 0
    steps = [x]

    flag = 0

    r = -grad_f(x)
    S = r

    while iterations < 1000:
        phi = lambda a: f(x + a * S)
        alpha = minimize_scalar(phi).x
        x_new = x + alpha * S

        r_new = -grad_f(x_new)
        if param == 1:
          beta = np.dot(r_new, r_new) / np.dot(r, r)
        elif param == 2:
          beta = max(0, np.dot(r_new, r_new - r)/np.dot(r, r))

        if flag:
            beta = 0

        S = r_new + beta * S

        iterations += 1
        steps.append(x_new)

        if (np.linalg.norm(x_new - x) < epsilon):
            if flag:
                x = x_new
                break
            else:
                flag = 1
                x = x_new
                r = r_new
                continue

        flag = 0

        x = x_new
        r = r_new

    return x, iterations, np.array(steps)

epsilon = 1e-6

print('Алгоритм Флетчера-Ривса')
starting_point = np.array([0.1, -1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 1)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

starting_point = np.array([0.1, -1.1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 1)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

starting_point = np.array([0, -1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 1)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

print('\nАлгоритм Полака-Рибьера')
starting_point = np.array([0.1, -1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 2)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

starting_point = np.array([0.1, -1.1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 2)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

starting_point = np.array([0, -1])
min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 2)
print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

print('\nвторой вариант работает лучше')
```
